```python
import os

def corpus_stats(file_path):
    """
    Compute vocabulary size and number of documents in a plain text corpus.
    Assumptions:
      - Each line = one document
      - Words are space-separated
    """
    vocab = set()
    num_docs = 0

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            words = line.strip().split()
            if not words:
                continue
            num_docs += 1
            vocab.update(words)

    vocab_size = len(vocab)
    return vocab_size, num_docs


def memory_requirement(vocab_size, num_docs):
    """
    Calculate memory requirement for a dense V x D matrix
    with 64-bit floats (8 bytes each).
    """
    bytes_needed = vocab_size * num_docs * 8
    mb = bytes_needed / (1024**2)
    gb = bytes_needed / (1024**3)
    tb = bytes_needed / (1024**4)
    return mb, gb, tb


def summarize_corpora(file_paths):
    """
    For each corpus file, compute stats and print summary table.
    """
    print("Corpus | Vocabulary | Documents | Memory (MB) | Memory (GB) | Memory (TB)")
    print("-" * 80)

    for file in file_paths:
        V, D = corpus_stats(file)
        mb, gb, tb = memory_requirement(V, D)
        print(f"{os.path.basename(file)} | {V:,} | {D:,} | {mb:,.2f} | {gb:,.2f} | {tb:,.3f}")


# Example usage
if __name__ == "__main__":
    # Replace with your actual corpus file paths
    files = [
        corpus_file
    ]
    summarize_corpora(files)
```
