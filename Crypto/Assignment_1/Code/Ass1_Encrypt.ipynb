```python
import random
import string
```

```python
key_size=5
key_file = "/mnt/Personal/Projects/Micro_Codes/Python/Crypto/key.txt"

input_file="/mnt/Personal/Projects/Micro_Codes/Python/Crypto/input.txt"
message_file="/mnt/Personal/Projects/Micro_Codes/Python/Crypto/message.txt"
output_file="/mnt/Personal/Projects/Micro_Codes/Python/Crypto/output.txt"
```

```python

def random_five_letters(size):
    return ''.join(random.choice(string.ascii_lowercase) for _ in range(size))

key = random_five_letters(key_size)
```

```python
print(key)
```

```python
def file_to_string(filename):
    try:
        with open(filename, 'r') as file:
            return file.read()
    except FileNotFoundError:
        raise FileNotFoundError(f"The file '{filename}' was not found")
    except IOError as e:
        raise IOError(f"Error reading file '{filename}': {str(e)}")
    
input_text = file_to_string(input_file)
```

```python
print(input_text)

print('\n\n',input_text.count(' ')," words")
```

```python
def preprocess_text(text):
    
    out=""
    
    for t in text:
        
        val = ord(t)
        
        if 65<= val <=90:
            val +=32
            
        if 97<= val <=122:
            out+=chr(val)
            
    return out

message = preprocess_text(input_text)
```

```python
print(message)
```

```python
def encrypt_text(key,text):
    
    size = len(key)
    out=""
    
    for i,t in enumerate(text):
        
        v1 = ord(key[(i%size)]) - 97
        v2 = ord(t) - 97
        
        dec = (v1+v2)%26 + 97
        
        out+=chr(dec)
        
    return out

encrypted = encrypt_text(key,message)
```

```python
print(encrypted)
```

```python
def write_file(filename, text):
    try:
        with open(filename, 'w') as file:
            file.write(text)
    except IOError as e:
        print(f"Error writing to file: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
write_file(message_file,message)
write_file(output_file,encrypted)
write_file(key_file,key)
```
