```python
import random

def generate_random_float():
    s= random.uniform(5, 10)
    print(s)
    return s

# Call the function to generate a random float number between 5 and 10
random_number = generate_random_float()
print(random_number)
```

```python
import time

def sleep_random():
    time.sleep(generate_random_float())

# Call the function to sleep for 5 seconds

import time

# Record the starting time
start_time = time.perf_counter()

sleep_random()

# Record the ending time
end_time = time.perf_counter()

# Calculate the elapsed time
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```

```python
import random
def func():
  random_number = random.randint(0, 4)
  #random_number=0
  #print(random_number)
  return 5//random_number
```

```python
def retry(max_retries=4,delay= 2):
    for attempt in range(max_retries):
        done=True
        try:
          func()
        except Exception as e:
          done=False
          print(f"\nAttempt {attempt} failed with error: {e}")
          if attempt < max_retries-1:
             print(f"\nRetrying in {delay} seconds...")
             time.sleep(delay)
        if (done):
          return
    raise RuntimeError(f"Function '{func.__name__}' failed after {max_retries} attempts")

retry()
```
